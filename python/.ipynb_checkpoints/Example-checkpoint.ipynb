```python
import numpy as np
from random import randint
from time import sleep
from uart_controller import UartController
from generic_controller import frame2uart, GenericController
from sklearn import datasets
from sklearn.preprocessing import normalize
from matplotlib import pyplot as plt
```

```python
try:
    del uc
except:
    pass

uc = UartController('/dev/ttyUSB0')
```

```python
uc.read()
uc.load_memory()
uc.flush()
```

```python
def print_switches():
    N = 32
    switches = np.zeros([N, N], dtype=np.byte)
    for i in range(N):
        for j in range(N):
            switches[i][j] = uc.get_switch(i, j)

    for i in range(N):
        print(",".join(["%2d"%w for w in switches[i]]))
        
        
def print_weights():
    N = 32
    weights = np.zeros([N, N+3], dtype=np.byte)
    for i in range(N):
        for j in range(N+3):
            weights[i][j] = uc.get_weight(i, j)

    for i in range(N):
        print(",".join(["%2d"%w for w in weights[i]]))
        

def print_weights_masked():
    N = 32
    weights = np.zeros([N, N+3], dtype=np.byte)
    for i in range(N):
        for j in range(N+3):
            weights[i][j] = uc.get_weight(i, j) if j >= N or uc.get_switch(i, j) != 0 else 0

    for i in range(N):
        print(",".join(["%2d"%w for w in weights[i]]))
```

```python
print_weights_masked()
```

```python
N = 32

for i in range(N):
    for j in range(N+3):        
#         uc.set_weight(i, j, 10 if j < N else 1 if j < N+2 else 3)
        uc.set_weight(i, j, 0 if j < N else 1)
        
        if j < N:
#             uc.set_switch(i, j, 4 if i < 4 else 2 if i < N-1 else 1)
            uc.set_switch(i, j, 0)    
            
#     uc.set_switch_output(i, 0 if i < N-3 else 1)
    uc.set_switch_output(i, 1)
    
uc.load_memory()
uc.flush()
```

```python
for i in range(32):
    uc.set_input_value(i, 2**32//2) # ustawienie generowania
    
uc.set_switch(0, 1, 4)  
uc.set_weight(0, 1, 15) 
uc.set_switch(2, 0, 4)  
uc.set_weight(2, 0, 15) 

uc.flush()
uc.set_output_value(0, 0) # reset liczników
uc.flush() 
sleep(0.1) #poczekanie na naładowanie
uc.queue_outputs('r') # odczyt liczników
uc.flush()

print([uc.get_output_value(i) for i in range(32)])
```

```python
uc.create_network(4, 8, 12, 9, 3, zero=True, add_bias=True)
uc.flush()
print_switches()
```

```python
X, Y = datasets.load_iris(return_X_y=True)
X = (normalize(X)*255).astype(int)
y_ohe = (
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1],
)

uc.set_batch_buffer_size(X.shape[0])
for i, (x, y) in enumerate(zip(X, Y)):
    uc.set_batch_entity(i, list(x), y_ohe[y])    
    
uc.flush()
```

```python
uc.set_learning_mode(0)
uc.set_learning_one(64)
uc.set_batch_loss(2**32-1)
uc.set_modify_probability(2**31)
uc.set_learning_actived(1)    
uc.flush()
# uc.flush()
print(uc.get_batch_loss(), uc.get_modify_probability())
for i in range(20):    
    uc.queue_config()
    uc.flush()
    print(uc.get_batch_loss(), uc.get_modify_probability())
    sleep(0.5)
uc.set_learning_actived(0)
uc.flush()
uc.load_memory()
uc.flush()
```

```python
uc.set_max_output_counter(256)

for x, y in zip(X, Y):
    for i in range(4):
        uc.set_input_value(i, 2**24 * x[i]) # ustawienie generowania
    uc.set_output_value(0, 0) # reset liczników
    uc.flush() 
    sleep(0.01) #poczekanie na naładowanie
    uc.queue_outputs('r') # odczyt liczników
    uc.flush()

    print([uc.get_output_value(i) for i in range(29, 32)], y_ohe[y])
```
